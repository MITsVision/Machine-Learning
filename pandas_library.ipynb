```python
import seaborn as sns

#allow plot to appear within notebook

%matplotlib inline
```

```python
sns.pairplot(data, x_vars=['TV','radio','newspaper'],y_vars='sales')
```

```python
sns.pairplot(data, x_vars=['TV','radio','newspaper'],y_vars='sales',size=8,aspect=0.7)
```
